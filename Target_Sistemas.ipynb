## Valor da varivel SOMA
---
    Ao final do processamento, qual ser o valor da varivel SOMA?

```python
INDICE = 13
SOMA = 0
K = 0

while K < INDICE:
    K = K + 1
    SOMA = SOMA + K

print(SOMA)
```

## Sequncia de Fibonacci
---
    Dado a sequncia de Fibonacci, onde se inicia por 0 e 1 e o prximo valor sempre ser a soma dos 2 valores anteriores;
    Escreva um programa na linguagem que desejar onde, informado um nmero, ele calcule a sequncia de Fibonacci;
    E retorne uma mensagem avisando se o nmero informado pertence ou no a sequncia.


```python
def fibonacci(numero):
    a, b = 0, 1
    while b < numero:
        a, b = b, a + b
    if b == numero:
        return f'{numero} pertence  sequncia de Fibonacci.'
    else:
        return f'{numero} no pertence  sequncia de Fibonacci.'

# Exemplo:
numero_informado = int(input("Digite um nmero: "))
resultado = fibonacci(numero_informado)
print(resultado)
```

## Lgica e prximo elemento
---
    Descubra a lgica e complete o prximo elemento:

      a) 1, 3, 5, 7, ___
      b) 2, 4, 8, 16, 32, 64, ____
      c) 0, 1, 4, 9, 16, 25, 36, ____
      d) 4, 16, 36, 64, ____
      e) 1, 1, 2, 3, 5, 8, ____
      f) 2,10, 12, 16, 17, 18, 19, ____

```python
def completa_padrao(padrao):
    if padrao == 'a':
        return [2 * i + 1 for i in range(5)]
    elif padrao == 'b':
        return [2**i for i in range(7)]
    elif padrao == 'c':
        return [i**2 for i in range(8)]
    elif padrao == 'd':
        return [i**2 for i in range(2, 6)]
    elif padrao == 'e':
        fibonacci = [1, 1]
        while len(fibonacci) < 8:
            fibonacci.append(fibonacci[-1] + fibonacci[-2])
        return fibonacci
    elif padrao == 'f':
        return [2, 10, 12, 16, 17, 18, 19, 20]

# Testando os padres
print(completa_padrao('a'))
print(completa_padrao('b'))
print(completa_padrao('c'))
print(completa_padrao('d'))
print(completa_padrao('e'))
print(completa_padrao('f'))
```

## Sala e interruptores
---
    Voc est em uma sala com trs interruptores, cada um conectado a uma lmpada em uma sala diferente.
    Voc no pode ver as lmpadas da sala em que est, mas pode ligar e desligar os interruptores quantas vezes quiser.
    Seu objetivo  descobrir qual interruptor controla qual lmpada.

Como descobriria, usando apenas duas idas at uma das salas das lmpadas, qual interruptor controla cada lmpada?

```python
def descobrir_lampadas_interruptores():
    # trs interruptores e trs lmpadas
    interruptores = [False, False, False]
    lampadas = [False, False, False]

    # Primeira ida
    # Ligando o primeiro interruptor
    interruptores[0] = True

    # Segunda ida
    # Ligando o segundo interruptor e desligando o primeiro
    interruptores[0] = False
    interruptores[1] = True

    # Observando o estado das lmpadas
    for i, estado in enumerate(lampadas):
        if estado or (not estado and interruptores[i]):
            print(f"O interruptor {i + 1} controla a lmpada {i + 1}")
        else:
            print(f"O interruptor {i + 1} no controla a lmpada {i + 1}")

descobrir_lampadas_interruptores()
```

## Inverso de caracteres de uma string
---
    Escreva um programa que inverta os caracteres de um string.

      IMPORTANTE:

        a) Essa string pode ser informada atravs de qualquer entrada de sua preferncia ou pode ser previamente definida no cdigo;
        b) Evite usar funes prontas, como, por exemplo, reverse;


```python
def inverter_string(s):
    return s[::-1]

# Exemplo
string_original = input("Digite uma string: ")  # Voc tambm pode definir uma string previamente no cdigo
resultado = inverter_string(string_original)
print(f"A string invertida : {resultado}")
```
